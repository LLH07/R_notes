- One of the most simple plotting tools in R is ```dot plot```.
- ```dot plot``` is called through <font color='blue'> ```plot()``` </font>.
- ```plot()```: takes in <font color='red'> 2 </font> equal length vectors.

````r
plotvector1 = 1:9
````

````r
plot(plotvector1, plotvector1)
````

可以看到 ```plot()``` 會把兩個 vectors 放到一個座標中。    
第一個 vector 丟給 <font color='red'> x 軸 </font>    
第二個 vector 丟給 <font color='red'> y 軸 </font>。  

補充:    
如果是用 R 的軟體畫圖，會出現 ```R Graphics: Device 2 (ACTIVATE)``` 的字樣。  
這表示現在這個視窗可以用來顯示圖。  

那如果你畫一個新的圖，舊的就會被取代。  
有時我們不想這麼做，就可以用 ```x11()``` 來產生一個新的視窗，且這時應該會顯示:   
```R Graphics: Device 3 (ACTIVATE)``` 與  
```R Graphics: Device 2 (inactive)```。

### Modify the graph:

````r
plotvector1 = 1:9
````

````r
plotvector2 = 9:1
````

````r
par(mfrow=c(1, 2), bg = 'grey', bty = 'n', cex = .75)
````

- ```mfrow```:讓視窗可以顯示多個圖。(而圖的排列方式為 ```row = 1```, ```column = 2```)
- ```bg```: 設定背景色。
- ```bty```: remove the box that normally surrounds our plot.
- ```cex```: make the size .75 of what it normally is.

````r
plot(plotvector1, plotvector2)
plot(plotvector1, plotvector1)
````

我在```R studio``` 可以成功把兩張圖畫成我們要的樣子，但在這裡不知為何無法。

提供幾個選項:
- ```xlab```, ```ylab```: 坐標軸的名稱。
- ```main```: 圖形的名稱。
- ```col```: 圖形顏色。
- ```pch = 3```: 用加號顯示。

````r
plot(plotvector1, plotvector2, xlab = "1:9", ylab = "9:1", main = "Plot of 1:9 versus 9:1", col = 'red', pch = 3)
````

### Histograms

使用之前的 ```porteinconc.csv``` 來畫直方圖。

proteinconc_header

目標:  
用第一個 column 的物質(```Nucleus```)，畫出內涵的 protein 與其 frequency:

````r
proteinconc = read.csv("proteinconc.csv", header = TRUE)
````

````r
class(proteinconc)
````

````r
colnames(proteinconc)
````

````r
colnames(proteinconc)[2] 
# this is what we want (recall that R starts from 1)
````

````r
hist(proteinconc[, 2], main = colnames(proteinconc)[2], xlab = "Concentration")
````

[參考資料: 'x' must be numeric](https://statisticsglobe.com/error-in-hist-default-x-must-be-numeric-in-r)

---
因為 R 的 histogram 得要輸入 numeric 的資料，但```proteinconc``` 是一個 ```data frame```。

[參考資料: convert data frame to numeric matrix](https://www.geeksforgeeks.org/convert-a-data-frame-into-a-numeric-matrix-in-r-programming-data-matrix-function/)

````r
proteinconc <- data.matrix(proteinconc)
````

````r
class(proteinconc)
````

````r
hist(proteinconc[, 2], main = colnames(proteinconc)[2], xlab = "Concentration")
````

前面的 histogram 的參數大概都跟 plot 一樣，但 histogram 也是有自己獨特的 parameters:

```R
par(mfrow=c(1, 2), cex = .75)
hist(proteinconc[, 1], breaks = 50)
hist(proteinconc[, 1], breaks = c(0, .5, .7, .8, .9, 1))
```

````r
hist(proteinconc[, 2], breaks = 50)
````

````r
hist(proteinconc[, 2], breaks = c(0, .5, .7, .8, .9, 1))
````

從圖形可以知道，```breaks``` 可以設定```x 座標``` 的單位與間隔。

看不懂@@
> ability to set frequency or density as the y-axis of your plot.  
Above, you see that hist() usually defaults to plotting frequency,  
but if you set bin limits that make bins of unequal length it changes to density (as it mathematically should).  
The function hist() can also be forced to plot density by adding the parameter "freq=FALSE".
